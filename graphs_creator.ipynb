```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Model': ['XGB', 'ARIMA', 'Prophet', 'RandomForest', 'SVR'],
    'MSE': [2678, 33853, 50413, 53.11, 1999],
    'MAE': [18.649, 155.22, 1571.92, 4.43, 5025.13],
    'R': [0.999, 0.987, 0.980, 0.999, 0.831]
}

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)

plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', cbar_kws={'label': 'Score'}, fmt='.4g', annot_kws={'size': 10})
plt.title('Dataset 2 Models Performance')
plt.show()
```
