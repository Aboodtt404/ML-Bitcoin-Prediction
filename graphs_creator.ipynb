```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Model': ['XGB', 'ARIMA', 'Prophet', 'RandomForest', 'SVR'],
    'MSE': [326572, 3363, 52708, 328380, 631568],
    'MAE': [262.91, 29.83, 141.54, 263.96, 384.33],
    'R': [0.300, 0.993, 0.887, 0.296, 0.354]
}

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)

plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', cbar_kws={'label': 'Score'}, fmt='.4g', annot_kws={'size': 10})
plt.title('Dataset 2 Models Performance')
plt.show()
```
